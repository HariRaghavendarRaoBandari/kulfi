```python
import pandas as pd
import numpy as np
import math as math
import matplotlib.pyplot as plt
%matplotlib inline
from pandas import Series

from patsy import dmatrices
import statsmodels.api as sm
from statsmodels.nonparametric.kde import KDEUnivariate
from statsmodels.nonparametric import smoothers_lowess
from sklearn import datasets, svm
from sklearn.ensemble import RandomForestClassifier

def paddedstr(num):
    if num<10:
        return '0'+str(num)
    else:
        return str(num)

filename='Abi/X'
tables=[0]
for filenumber in range(1, 25):
    curT=pd.read_csv('Abi/X'+paddedstr(filenumber),sep=' ',header=None,prefix='X')
    curT=curT.drop(['X0'],axis=1)
    for col in range(1,721):
        if (col%5!=1):
            curT=curT.drop(['X'+str(col)],axis=1)
        else:
            curT=curT.rename(columns={'X'+str(col):'X'+str(col/5+1)})
    curT['index']=np.arange(2016)
    tables.append(curT)    
```

```python
def updateAxis(maximumy,miniy=0):
    x1,x2,y1,y2 = plt.axis()
    if maximumy>0:
        plt.axis((x1,x2,miniy,maximumy))    
def plotDays(maximumy,gap=5000,alp=1):
    if maximumy>0:
        bound=maximumy
    else: 
        bound=10000000;
    for i in range(1,8):
        plt.plot([i*288]*(bound/gap),range(0,(bound/gap)*gap,gap),'rx',markersize=2,alpha=alp)
def plotCurve(serve,maximumy):
    plt.figure(figsize=(18,4));
    for i in serve:
        #i.plot()
        plt.plot(range(0,len(i)), i);        
    updateAxis(maximumy)
    plotDays(maximumy,100000,0.5)
def plotCurveTableData(atts,colname,maximumy):
    serve=[];
    for i in atts:
        serve.append(tables[i][colname])
    plotCurve(serve,maximumy)
def smoothedCurve(serve,interval):    
    output=[];
    for i in range(0,len(serve)):
        ans=0;
        if i-interval<=0:
            for j in range(0,i+interval+1):
                ans=ans+serve[j];
            ans=ans/(i+interval+1);
        elif i+interval+1>=len(serve):
            for j in range(i-interval,len(serve)):
                ans=ans+serve[j];
            ans=ans/(len(serve)-(i-interval));
        else:
            for j in range(i-interval,i+interval+1):
                ans=ans+serve[j];
            ans=ans/(interval*2+1);
        output.append(ans);
    return output       
        
def plotCurveSmoothedTableData(atts,colname,maximumy,interval=10):
    serve=[];
    for i in atts:
        serve.append(smoothedCurve(tables[i][colname],interval))
    plotCurve(serve,maximumy)
def plotScatter(serve,maximumy):
    count=0;
    numineachrow=3;
    for i in serve:
        count=count+1;
        if count%numineachrow==1:
            plt.figure(figsize=(18,4));
        plt.subplot('1'+str(numineachrow)+str((count-1)%numineachrow+1), axisbg="#DBDBDB")
        plt.plot(range(0,len(i)), i, '.', markersize=1);
        plotDays(maximumy)
        updateAxis(maximumy)
def plotScatterTableData(atts,colname, maximumy):
    serve=[];
    for i in atts:
        serve.append(tables[i][colname])
    plotScatter(serve,maximumy)

plotCurveSmoothedTableData([1,2,3,4,5],'X1',600005)
plotCurveTableData([1,2,3,4,5],'X1',600005)
plotScatterTableData([1,2,3,4,5],'X1',600005)
```

```python
def splitSignalCondition(serve):
    curv=serve[0];
    i=0;
    jump=[[],[]];
    cont=[]
    avg=np.average(serve)
    while i<=len(serve):
        cont.append(curv);
        lastv=curv;
        i=i+1;
        if i>=len(serve):
            break;
        curv=serve[i];        
        if (curv>lastv*2) & (curv-lastv>avg*2):
            jump[0].append(i)
            delta=curv-lastv
            jump[1].append(delta)
            while (True):
                cont.append(curv-delta);
                i=i+1;
                if i>=len(serve):
                    break;
                curv=serve[i];        
                if not((curv>lastv*2) & (curv-lastv>avg*2)):
                    break;                                
    return [cont,jump]        
```

```python
def splitSignalMedian(serve,l=4):
    serve=serve.values.tolist()
    tmp=serve[0:l]
    avg=np.average(serve)
    jump=[[],[]];
    for i in range(l,len(serve)-l):
        tmp.append(np.median(serve[i-l:i+l]))
        if (-tmp[i]+serve[i])>avg*2:
            jump[0].append(i)
            jump[1].append(serve[i])  
    for i in range (len(serve)-l,len(serve)):
        tmp.append(serve[i])
    return [tmp,jump] 
def printsignal(serve,cont,jump,maximumy,miniy=0):
    plt.figure(figsize=(18,4));
    plt.plot(range(0,len(cont)), serve, 'b-', markersize=1);
    plotDays(0)
    updateAxis(maximumy)
    plt.figure(figsize=(18,4));
    plt.plot(range(0,len(cont)), cont, 'b-', markersize=1);
    plt.plot(jump[0], jump[1], 'g*', markersize=5);
    plotDays(0)
    updateAxis(maximumy)    
    plt.figure(figsize=(18,4));
    plt.plot(range(0,len(cont)), serve-cont, 'b-', markersize=1);
    plotDays(0)
    updateAxis(maximumy,miniy)
```

```python
def sqr(a):
    return a*a;
def getloss(serve, predict):
    regret=[0,0,0,0];
    for i in range(0,len(serve)):
        regret[0]=regret[0]+math.log(abs(serve[i]-predict[i])+1);
        regret[1]=regret[1]+sqr(math.log(abs(serve[i]-predict[i])+0.001));
        if predict[i]>=serve[i]:
            regret[2]=regret[2]+predict[i];
        else:
            regret[2]=regret[2]+serve[i]+(serve[i]-predict[i])*10;
        regret[3]=regret[3]+abs(serve[i]-predict[i]);
    regret[2]=regret[2]/sum(serve);
    regret[0]=regret[0]/len(serve);
    regret[1]=regret[1]/len(serve);
    regret[3]=regret[3]/len(serve);
    return regret     
```
